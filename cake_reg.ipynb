```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets

#plot library
%matplotlib inline
import matplotlib.pyplot as plt
import scikitplot as plt1 #pip install scikitplot
```

```python
feature=pd.read_csv("cake.csv",header=None,usecols=range(1,9),skiprows=1)
feature.shape
```

```python
fea=np.array(feature)
fea.shape
```

```python
ty=pd.read_csv("cake.csv",header=None,usecols=range(0,1),skiprows=1)
ty.shape
```

```python
ty=np.array(ty)
ty
```

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
le.fit(ty)
ty=le.transform(ty)
ty=np.array(ty)
print(ty)
```

```python
X=feature
Y=ty
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=12)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
```

```python
reg=reg.fit(X_train,Y_train)
print("Slope:", reg.coef_[0])
```

```python
print("Intercept:",reg.intercept_)
```

```python
y_pred_train=reg.predict(X_train)
y_pred_test=reg.predict(X_test)

print("Actual Train output is", Y_train)
print("Predicit Train Output is", y_pred_train)
print(" ")
print("Actual Train output is", Y_test)
print("Predicit Train Output is", y_pred_test)
```

```python
plt.scatter(X_train[:,0],Y_train,color="black")
plt.plot(X_train[:,0],y_pred_train,color="blue",linewidth=3)

plt.show()
```

```python
plt.scatter(X_test[:,0],Y_test,color="black")
plt.plot(X_test[:,0],y_pred_test,color="blue",linewidth=3)

plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score

print("Mean squared error: %.2f" % mean_squared_error(Y_test,y_pred_test))
```

```python
print("R2 sore: %.2f" % r2_score(Y_test,y_pred_test))
```
